```python
#once the desired coordinates are figured out choose variables to assign the contents of the coordinates into
#function below is specific to the pdf file for revation since those coordinates reflect what is in the pdf


def pdfscrape(pdf):
    # Extract each relevant information individually
    name              = pdf.pq('LTTextLineHorizontal:overlaps_bbox("148.896, 733.757, 220.15, 750.683")').text()
    date              = pdf.pq('LTTextLineHorizontal:overlaps_bbox("148.896, 754.493, 206.832, 771.419")').text()
    starttime         = pdf.pq('LTTextLineHorizontal:overlaps_bbox("174.24, 656.552, 273.427, 670.874")').text()
    endtime           = pdf.pq('LTTextLineHorizontal:overlaps_bbox("174.24, 631.976, 273.427, 646.298")').text()
    onlinetime        = pdf.pq('LTTextLineHorizontal:overlaps_bbox("200.352, 572.072, 238.191, 586.394")').text()
    breaktime         = pdf.pq('LTTextLineHorizontal:overlaps_bbox("200.352, 461.48, 238.191, 475.802")').text()
    inboundcalls      = pdf.pq('LTTextLineHorizontal:overlaps_bbox("200.352, 332.456, 211.047, 346.778")').text()
# Combined all relevant information into single observation
# creating a pandas data frame to assign the variables above to specific columns labeled after the variables
    page = pd.DataFrame({
                         'name': name,
                         'date': date,
                         'starttime': starttime,
                         'endtime': endtime,
                         'onlinetime': onlinetime,
                         'breaktime': breaktime,
                         'inboundcalls': inboundcalls
                       }, index=[0])
    return(page)
```

```python
#the code below gets the number of pages in the pdf file and
pagecount = pdf.doc.catalog['Pages'].resolve()['Count']


master = pd.DataFrame()
#the for loop is now going to go into every page in the pdf
for p in range(pagecount):
    #loading that speficic page from the pdf
    pdf.load(p)
    #scraping the contents of that specific page
    page = pdfscrape(pdf) 
    #appending everything to the master pandas dataframe so everything is in one pandas frame
    master = master.append(page, ignore_index=True)
#exporting to a csv file    
master.to_csv('output.csv', index = False)
```
